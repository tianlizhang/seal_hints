```python
lst = []

for ii in range(len(paths)):
    fr = open(paths[ii], 'r')
    lines = fr.readlines()
    fr.close()

    tsp = int(paths[ii].split('/')[-1][2:-4])
    for line in tqdm(lines, desc=f'{paths[ii]}'):
        parts = line.split('\t')
        if len(parts) == 2:
            try:
                src, tgt = int(parts[0]), int(parts[1])
                lst.append((src, tgt, tsp))
            except:
                continue
len(lst)
```

```python
src = [item[0] for item in lst]
tgt = [item[1] for item in lst]
tsp = [item[2] for item in lst]

papers = pd.DataFrame({'source': src, 'target': tgt, 'time': tsp})
papers = papers.sort_values(by=['time', 'source', 'target'])
papers
```

## cut

```python
cut_step = (papers['time'].max() - papers['time'].min()) / 20
cut_step
```

```python
papers['new_time'] = papers['time'].map(lambda x: int((x-papers['time'].min())/cut_step))
papers
```
